Get the list of unique tracks from spotify file. Ouput needs 'song_name', 'artist_name'and 'album_name' (known as release in the database) in order to match

```python
import json

with open('YourLibrary.json', encoding="utf-8") as json_data:
    data = json.load(json_data)
    bensLibrary = pd.DataFrame(data['tracks'])

# only keep the needed columns for sorting
sortingDf = bensLibrary[['track', 'artist', 'album']]
```

```python
# escape all single quotes
escapedDf = sortingDf.replace("'", "''", regex=True)
escapedDf.head()
```

```python


connection = sqlite3.connect('track_metadata.db')

# for every song in bens library make a sql query and retrieve the track id
escapedDf.head()
escapedDf['id'] = escapedDf.apply(lambda x: pd.read_sql_query(f"SELECT track_id FROM songs WHERE title='{x[0]}' AND artist_name='{x[1]}' AND release='{x[2]}'", connection), axis=1)
connection.close()
escapedDf.head()
```
